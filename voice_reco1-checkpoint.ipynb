# 输入两段音频，判断这两个音频是不是一个人的音色

```python
import librosa  
from sklearn.metrics.pairwise import cosine_similarity  
  
def extract_mfccs(audio_path, sr=22050, n_mfcc=13):  
    # 加载音频文件  
    y, sr = librosa.load(audio_path, sr=sr)  
      
    # 提取MFCC特征  
    mfccs = librosa.feature.mfcc(y=y, sr=sr, n_mfcc=n_mfcc)  
      
    return mfccs  
  
def compare_audios(audio_path1, audio_path2):  
    # 提取两个音频文件的MFCC特征  
    mfccs1 = extract_mfccs(audio_path1)  
    mfccs2 = extract_mfccs(audio_path2)  
    
    print(mfccs1.shape)
    print("----")
    print(mfccs2.shape)
      
    # 确保两个MFCC矩阵有相同的列数（即特征数量）  
    # 这里我们选择截断较长的音频的MFCC特征  
    min_mfccs = min(mfccs1.shape[1], mfccs2.shape[1])  
    mfccs1 = mfccs1[:, :min_mfccs]  
    mfccs2 = mfccs2[:, :min_mfccs]  
    
    print(mfccs1.shape)
    print("----")
    print(mfccs2.shape)
      
    # 计算两个MFCC特征矩阵之间的余弦相似度  
    # 注意：cosine_similarity 期望输入是二维数组  
    similarity = cosine_similarity(mfccs1, mfccs2)  
      
    # 输出相似度分数  
    print(f"Similarity between the two audio files: {similarity[0][0]}")  
    
    

  
# 使用函数  
audio_file1 = '明星声音库/男声-蔡徐坤（手机内录）.wav'  
audio_file2 = '明星声音库/男声-蔡徐坤.wav'  
compare_audios(audio_file1, audio_file2)
```

```python
import os
import librosa
import numpy as np
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader, Dataset
from torchvision import transforms

# 定义声音数据集类
class SoundDataset(Dataset):
    def __init__(self, male_dir, female_dir, transform=None, max_length=8000):
        self.male_files = [os.path.join(male_dir, f) for f in os.listdir(male_dir)]
        self.female_files = [os.path.join(female_dir, f) for f in os.listdir(female_dir)]
        self.files = self.male_files + self.female_files
        self.labels = [0] * len(self.male_files) + [1] * len(self.female_files)
        print(self.files)
        self.transform = transform
        self.max_length = max_length

    def __len__(self):
        return len(self.files)

    def __getitem__(self, idx):
        file_path = self.files[idx]
        waveform, sample_rate = librosa.load(file_path, sr=None)
        # 截断或填充到相同的长度
        if len(waveform) < self.max_length:
            waveform = np.pad(waveform, (0, self.max_length - len(waveform)))
        else:
            waveform = waveform[:self.max_length]
        if self.transform:
            waveform = self.transform(waveform)
        label = self.labels[idx]
        return waveform, label

# 定义声音转换函数
class ToTensor(object):
    def __call__(self, waveform):
        return torch.tensor(waveform, dtype=torch.float32)

# 定义简单的CNN模型
class GenderClassifier(nn.Module):
    def __init__(self):    # 定义部分
        super(GenderClassifier, self).__init__()
        self.conv1 = nn.Conv1d(in_channels=1, out_channels=32, kernel_size=3, stride=1, padding=1)
        self.conv2 = nn.Conv1d(in_channels=32, out_channels=64, kernel_size=3, stride=1, padding=1)
        self.conv3 = nn.Conv1d(in_channels=64, out_channels=128, kernel_size=3, stride=1, padding=1)
        self.fc1 = nn.Linear(128 * 1000, 128)
        self.fc2 = nn.Linear(128, 2)
        self.relu = nn.ReLU()
        self.maxpool = nn.MaxPool1d(kernel_size=2, stride=2)

    def forward(self, x):    # 应用部分
#         print(x.size())
#         print(x)
        batch_size = x.size(0)
        # 获取音频的长度
        audio_length = x.size(1)
        # 重塑输入张量，使其适应不同的批次大小和音频长度
        x = x.view(batch_size, 1, audio_length)
        
        x = self.relu(self.conv1(x))
        x = self.maxpool(x)
        x = self.relu(self.conv2(x))
        x = self.maxpool(x)
        x = self.relu(self.conv3(x))
        x = self.maxpool(x)
        x = x.view(-1, 128 * 1000)
        x = self.relu(self.fc1(x))
        x = self.fc2(x)
#         print(x)
#         print("____x____")
        return x

# 设置训练参数
batch_size = 2
learning_rate = 0.00001
num_epochs = 5

# 准备数据
transform = transforms.Compose([ToTensor()])
train_dataset = SoundDataset(male_dir='voicedata/easyyy', female_dir='voicedata/easzzz', transform=transform)
train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True)

# 初始化模型、损失函数和优化器
model = GenderClassifier()
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=learning_rate)

# 在训练循环之外定义评估函数
def evaluate(model, dataloader):
    model.eval()  # 将模型设为评估模式
    correct = 0
    total = 0
    with torch.no_grad():  # 在评估模式下不需要计算梯度
        for waveforms, labels in dataloader:
            outputs = model(waveforms)
            _, predicted = torch.max(outputs, 1)  # 获取预测结果中的最大值及其对应的索引
            total += labels.size(0)  # 累加样本数量
            correct += (predicted == labels).sum().item()  # 累加预测正确的样本数量
    accuracy = correct / total  # 计算准确度
    return accuracy

# 在每个epoch结束后评估模型并打印准确度
for epoch in range(num_epochs):
    for i, (waveforms, labels) in enumerate(train_loader):
#         print("len:" + str(len(train_loader)))
        total_steps = len(train_loader)  # 将total_steps定义在循环内部
        outputs = model(waveforms)
        loss = criterion(outputs, labels)
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()
        
        print(i)
        if (i+1) % 1 == 0:
            print(f'Epoch [{epoch+1}/{num_epochs}], Step [{i+1}/{total_steps}], Loss: {loss.item():.4f}')
    
    # 在每个epoch结束后评估模型并打印准确度
    accuracy = evaluate(model, train_loader)
    print(f'Epoch [{epoch+1}/{num_epochs}], Accuracy: {accuracy:.4f}')

# 保存模型
torch.save(model.state_dict(), 'gender_classifier.pth')
```

```python
import os
import librosa
import numpy as np
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader, Dataset
from torchvision import transforms

# 定义声音数据集类
class SoundDataset(Dataset):
    def __init__(self, male_dir, female_dir, transform=None, max_length=8000, n_mfcc=20):
        self.male_files = [os.path.join(male_dir, f) for f in os.listdir(male_dir)]
        self.female_files = [os.path.join(female_dir, f) for f in os.listdir(female_dir)]
        self.files = self.male_files + self.female_files
        self.labels = [0] * len(self.male_files) + [1] * len(self.female_files)
        self.transform = transform
        self.max_length = max_length
        self.n_mfcc = n_mfcc

    def __len__(self):
        return len(self.files)

    def __getitem__(self, idx):
        file_path = self.files[idx]
        waveform, sample_rate = librosa.load(file_path, sr=None)
        # 计算MFCC特征
        mfcc = librosa.feature.mfcc(y=waveform, sr=sample_rate, n_mfcc=self.n_mfcc)
        # 将MFCC特征转换为张量
        mfcc = torch.tensor(mfcc, dtype=torch.float32)
        # 截断或填充到相同的长度
        if mfcc.shape[1] < self.max_length:
            mfcc = torch.cat((mfcc, torch.zeros((self.n_mfcc, self.max_length - mfcc.shape[1]))), dim=1)
        else:
            mfcc = mfcc[:, :self.max_length]
        if self.transform:
            mfcc = self.transform(mfcc)
        label = self.labels[idx]
        return mfcc, label

# 定义声音转换函数
class ToTensor(object):
    def __call__(self, mfcc):
        return mfcc

# 定义简单的CNN模型
class GenderClassifier(nn.Module):
    def __init__(self, input_size, num_classes):    # 定义部分
        super(GenderClassifier, self).__init__()
        self.conv1 = nn.Conv1d(in_channels=input_size, out_channels=32, kernel_size=3, stride=1, padding=1)
        self.conv2 = nn.Conv1d(in_channels=32, out_channels=64, kernel_size=3, stride=1, padding=1)
        self.conv3 = nn.Conv1d(in_channels=64, out_channels=128, kernel_size=3, stride=1, padding=1)
        self.fc1 = nn.Linear(128 * 1000, 128)
        self.fc2 = nn.Linear(128, num_classes)
        self.relu = nn.ReLU()
        self.maxpool = nn.MaxPool1d(kernel_size=2, stride=2)

    def forward(self, x):    # 应用部分
        batch_size = x.size(0)
        x = x.view(batch_size, x.size(1), -1)
        
        x = self.relu(self.conv1(x))
        x = self.maxpool(x)
        x = self.relu(self.conv2(x))
        x = self.maxpool(x)
        x = self.relu(self.conv3(x))
        x = self.maxpool(x)
        x = x.view(-1, 128 * 1000)
        x = self.relu(self.fc1(x))
        x = self.fc2(x)
        return x

# 设置训练参数
batch_size = 2
learning_rate = 0.00001
num_epochs = 5
max_length = 8000
n_mfcc = 20

# 准备数据
transform = transforms.Compose([ToTensor()])
train_dataset = SoundDataset(male_dir='voicedata/aaa', female_dir='voicedata/uuu', transform=transform, max_length=max_length, n_mfcc=n_mfcc)
train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True)

# 初始化模型、损失函数和优化器
model = GenderClassifier(input_size=n_mfcc, num_classes=2)
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=learning_rate)

# 在训练循环之外定义评估函数
def evaluate(model, dataloader):
    model.eval()  
    correct = 0
    total = 0
    with torch.no_grad():  
        for waveforms, labels in dataloader:
            outputs = model(waveforms)
            _, predicted = torch.max(outputs, 1)  
            total += labels.size(0)  
            correct += (predicted == labels).sum().item()  
    accuracy = correct / total  
    return accuracy

# 在每个epoch结束后评估模型并打印准确度
for epoch in range(num_epochs):
    for i, (waveforms, labels) in enumerate(train_loader):
        total_steps = len(train_loader)  
        outputs = model(waveforms)
        loss = criterion(outputs, labels)
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()
        
        if (i+1) % 1 == 0:
            print(f'Epoch [{epoch+1}/{num_epochs}], Step [{i+1}/{total_steps}], Loss: {loss.item():.4f}')
    
    accuracy = evaluate(model, train_loader)
    print(f'Epoch [{epoch+1}/{num_epochs}], Accuracy: {accuracy:.4f}')

# 保存模型
torch.save(model.state_dict(), 'gender_classifier.pth')
```
